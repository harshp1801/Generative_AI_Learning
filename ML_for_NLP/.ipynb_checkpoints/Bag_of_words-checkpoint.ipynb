```python
import nltk
import pandas as pd
import numpy as np
import csv
nltk.download('stopwords')
import re
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
stop_words = stopwords.words('english') #mimp if not specified it throws lazycorpusloader() error
lemmatizer = WordNetLemmatizer()
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(max_features = 2500)
binaryvect = CountVectorizer(max_features = 2500,binary = True)
path = '/home/harsh-ideapad/Gen_AI_Learning/ML_for_NLP/Bag_of_words/Dataset/spam.csv'
```

```python
target = []
text = []
with open(path, encoding='utf-8', errors='replace', newline='') as f:
    reader = csv.reader(f)
    for row in reader:
        target.append(row[0])
        row.pop(0)
        text.append(''.join(i.lower() for i in row))
```

```python
data = zip(text[1:],target[1:])
df = pd.DataFrame(data,columns = ['text','spam/ham'])
```

```python
df
```

```python
# Data Cleaning And Preprocessing
```

```python
cleaned = []
for sentence in df['text']:
    cleaned_text = []
    processed_sent = re.sub('[^a-zA-Z]',' ',sentence) 
    for word in processed_sent.split():
        if word not in stop_words:
            cleaned_text.append(lemmatizer.lemmatize(word))
    cleaned.append(' '.join(i for i in cleaned_text))
```

```python
cleaned
```

```python
Xbow = vectorizer.fit_transform(cleaned)
Xbow.shape
```

```python
Xbinbow = binaryvect.fit_transform(cleaned)
Xbinbow.shape
```

```python
vectorizer.vocabulary_ # Here the key representts the vocabulary and value represents the index
```
