```python
lst = ["I am a good boy","The glass of milk","The cup of tea","I am a good developer","your videos are good","I will become the best data scientist"]
```

```python
len(lst)
```

```python
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Embedding
```

```python
vocab_size = 1000
```

```python
lst
```

```python
one_hot_repre = [one_hot(word,vocab_size)for word in lst]
one_hot_repre 
```

    This numbers represent the index where the value is 1 for that word like 
    In sentence 1 am is represented as a sparse matrix with 1 at index 72


```python
sent_length = 8
embedded_pre = pad_sequences(one_hot_repre,padding = "pre",maxlen = sent_length)
```

```python
embedded_pre
```

```python
embedded_post = pad_sequences(one_hot_repre,padding = "post",maxlen = sent_length)
```

```python
embedded_post
```

```python
model = Sequential()
```

```python
dim = 10
model.add(Embedding(vocab_size,dim,input_length = sent_length))
model.build(input_shape=(None,8))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_pre) # Every word is represented by dimension 10
```
